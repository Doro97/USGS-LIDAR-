```python
import os
from glob import glob
import matplotlib.pyplot as plt
import rasterio as rio
from rasterio.plot import plotting_extent
from rasterio.plot import show
import geopandas as gpd
import earthpy as et
import earthpy.spatial as es
import earthpy.plot as ep
```

```python
#reading the iowa raster file
fp = r'/content/iowa.tif'
```

```python
dataset=rio.open('/content/iowa.tif')
dataset
```

```python
#raster file attributes
print('Number of bands:' ,dataset.count)
print('Image Resoultion:',dataset.height, dataset.width)
print('CRS:',dataset.crs)
```

# Visualization

```python

plt.figure(figsize=(10, 6))
show(dataset)
```


The values on the x and y axis represent the latitude and longitude coordinates

```python
plt.figure(figsize=(10, 6))
plt.imshow(dataset.read(1), cmap='pink')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
show((dataset, 1), transform=dataset.transform, cmap='viridis')
```

## Plotting raster file with contours

```python
fig, ax = plt.subplots(1, figsize=(12, 12))
show((dataset, 1), ax=ax, contour=True, contour_label_kws={})
```

## Plotting a histogram

```python
plt.figure(figsize=(12, 8))
from rasterio.plot import show_hist
show_hist(
    dataset, bins=50, lw=0.0, stacked=False, alpha=0.3,
   histtype='stepfilled', title="Histogram of the raster")
```

```python
fig, (axrgb, axhist) = plt.subplots(1, 2, figsize=(14,7))
show((dataset), cmap='Greys_r', contour=True, ax=axrgb)
show_hist(dataset, bins=50, histtype='stepfilled',
          lw=0.0, stacked=False, alpha=0.3, ax=axhist)
plt.show()
```
